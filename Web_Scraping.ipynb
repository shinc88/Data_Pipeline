### Import library

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/List_of_cities_in_the_European_Union_by_population_within_city_limits"

headers = {'Accept-Language': 'en-US,en;q=0.8'}
response = requests.get(url, headers = headers)
response.status_code

soup = BeautifulSoup(response.content, "html.parser")
soup
```

### Locate data from JSON file 

```python
#select the cities table
table = soup.find_all('tbody')[0]
table
```

```python
#extract table - 94 cities(rows)
rows = table.find_all('tr')
rows
```

```python
#extract city name 
city_name = rows[1].find_all('td')[1].get_text()
city_name
```

```python
#extract country name
countries_name = rows[1].find_all('td')[2].get_text()
countries_name
```

```python
#extract population
population = rows[1].find_all('td')[3].get_text()
population
```

```python
#extract population timestamp
population_timestamp = rows[1].find_all('td')[4].get_text()
population_timestamp
```

### Create a loop and store data in a data frame

```python
city=[]
country=[]
population=[]
population_timestamp=[]

for i in range(1,len(rows)):
    city.append(rows[i].find_all('td')[1].get_text())
    country.append(rows[i].find_all('td')[2].get_text())
    population.append(rows[i].find_all('td')[3].get_text())
    population_timestamp.append(rows[i].find_all('td')[4].get_text())

cities_df = pd.DataFrame({"city":city,"country":country,"population":population,'population_timestamp':population_timestamp})
cities_df
```

### Bonus - data cleaning

```python
cities_df['country'] = cities_df.country.str.replace('\n','')
cities_df['population'] = cities_df.population.str.replace('\n','')
cities_df['population'] = cities_df.population.str.replace(',','')
cities_df['population_timestamp'] = cities_df.population_timestamp.str.replace('\n','')
cities_df.population_timestamp = pd.to_datetime(cities_df.population_timestamp)
cities_df.population = pd.to_numeric(cities_df.population)
```
