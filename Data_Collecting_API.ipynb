### Create a loop and store data in a data frame

```python
weather_cities_df = {"city":[],
                     "longitude":[],
                     "latitude":[], 
                     "temperature":[],
                     "wind_speed":[],
                     "time_record":[],
                     "weather_condition":[]}

for i in weather_json['list']:
    weather_cities_df['city'].append(weather_json['city']['name'])
    weather_cities_df['longitude'].append(weather_json['city']['coord']['lon'])
    weather_cities_df['latitude'].append(weather_json['city']['coord']['lat'])
    weather_cities_df['temperature'].append(i['main']['temp'])
    weather_cities_df['wind_speed'].append(i['wind']['speed'])
    weather_cities_df['time_record'].append(i['dt_txt'])
    weather_cities_df['weather_condition'].append(i['weather'][0]['main'])

weather = pd.DataFrame(weather_cities_df)
weather
```

### Function for multiple cities

```python
cities = ['Berlin','Madrid','Paris']

def get_weather_loop(cities):
    
    API_key = 'YOUR_API_KEY'

    weather_cities = {"city":[],
                      "country":[],
                      "longitude":[],
                      "latitude":[], 
                      "temperature":[],
                      "wind_speed":[],
                      "time_record":[],
                      "weather_condition":[]
                     }
    
    for city in cities:
        url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}&units=metric"
        response = requests.get(url)
        weather_multi_city_json = response.json()
        
        for i in weather_multi_city_json['list']:
                weather_cities['city'].append(weather_multi_city_json['city']['name'])
                weather_cities['country'].append(weather_multi_city_json['city']['country'])
                weather_cities['longitude'].append(weather_multi_city_json['city']['coord']['lon'])
                weather_cities['latitude'].append(weather_multi_city_json['city']['coord']['lat'])
                weather_cities['temperature'].append(i['main']['temp'])
                weather_cities['wind_speed'].append(i['wind']['speed'])
                weather_cities['time_record'].append(i['dt_txt'])
                weather_cities['weather_condition'].append(i['weather'][0]['main'])
                
    return pd.DataFrame(weather_cities) 

weather = get_weather_loop(cities)
weather
```

### Bonus - data cleaning

```python
weather.time_record = pd.to_datetime(weather.time_record)
weather.info()
```
